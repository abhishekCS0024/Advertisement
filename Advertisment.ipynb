```python
x1=df_encoded.drop(columns='Clicked on Ad')
```

```python
x2=df_encoded['Clicked on Ad']
```

```python
from sklearn.model_selection import train_test_split
# train means 80% of data & test means 20% of data
# x means independent feature & y means dependent or target features
xtrain,xtest,ytrain,ytest = train_test_split(x1,x2,test_size=0.3,random_state=32)
xtrain
```

```python
from sklearn.linear_model import LogisticRegression
Model=LogisticRegression()
```

```python
Model.fit(xtrain,ytrain)# We fit our 80% of data into the model
```

```python
p=Model.predict(xtest)
p
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
accuracy_score(p,ytest)*100
```
